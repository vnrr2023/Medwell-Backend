```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
```

```python
# Try with 'ISO-8859-1' or 'cp1252' encoding
document = TextLoader("disease_data.txt", encoding="ISO-8859-1").load()

text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=1000,
    chunk_overlap=750
)
```

```python
len(document),type(document)
```

```python
texts = text_splitter.split_documents(document)
```

```python
type(texts)
```

```python
len(texts)
```

```python
len(texts[2].page_content)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = "med_stack_chatbot_db"
collection_name="remedies"
```

```python
vectordb = Chroma.from_documents(
    documents=texts,
    embedding=embeddings,
    persist_directory=persist_directory,
    collection_name=collection_name
)
```

```python
# retriever
retriever = vectordb.as_retriever()
```

```python
# llm from groq
llm = ChatGroq(
    model="llama-3.1-70b-versatile",
    temperature=0.2
)
```

```python
# create a qa chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
```

```python
# invoke the qa chain and get a response for user query
query = "what are the symptoms of anal itching?"
response = qa_chain.invoke({"query": query})
```
